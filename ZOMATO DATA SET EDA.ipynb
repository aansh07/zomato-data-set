```python
df2=pd.read_excel('Country-Code.xlsx')
```

```python
df2.head()
```

```python
final_df=pd.merge(df,df2,on='Country Code',how="left") #how tells u on which dataset df u want to focus like join left table or right table
```

```python
final_df.Country.value_counts() #value_counts tells how many time country has occur in the dataframe
```

```python
country_name=final_df.Country.value_counts().index
```

```python
country_values=final_df.Country.value_counts().values
```

```python
#pie chart=only for top 3 countries that uses zomato
plt.pie(country_values[:3],labels=country_name[:3],autopct='%1.9f%%') #autopct gives you the 
                                                                     #percentage on the pie chart.
```

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

### Observation
1.When rating is between 4.5 and 4.9 it is excellent.
2.When rating is between 4.0 and 4.5 it is good.


```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams["figure.figsize"]=(12,10)
sns.barplot(data=ratings,x='Aggregate rating',y='Rating Count',hue="Rating color")
```

# observation
     1.Not rated count is very high.
     2.Maximum number of  rating are between  2.5 to 3.4

```python
#countplot--> to implement the categorical values.
import seaborn as sns
sns.countplot(x="Rating color",data=ratings) # this show how frequently they are display in the dataframe.
```
